```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Heirarchical Clustering

```python
f = "../assignment-2-ksr-0011/Data/SMAI-Dataset-hc-dataset/new_customers.csv"
data = pd.read_csv(f)
data = data.drop(['CustomerID'], axis=1)
data.head()
```

```python
# Coverting male and female to 0 and 1
data['Genre'] = data['Genre'].replace({'Male': 0, 'Female': 1})
data.head()
```

```python
data.shape
```

```python

heirarchical_single = linkage(data, method='single')
heirarchical_complete = linkage(data, method='complete')
heirarchical_average = linkage(data, method='average')
heirarchical_ward = linkage(data, method='ward') 

c3c4 = data.iloc[:, [2,3]].values
c1c3 = data.iloc[:, [0,2]].values
c1c2 = data.iloc[:, [0,1]].values
heirarchical_ward_c3c4 = linkage(c3c4, method='ward')
heirarchical_ward_c1c3 = linkage(c1c3, method='ward')
heirarchical_ward_c1c2 = linkage(c1c2, method='ward')


plt.figure(figsize=(15, 9))
plt.title("Single Linkage Dendogram")
dendrogram(heirarchical_single)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Complete Linkage Dendogram")
dendrogram(heirarchical_complete)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Average Linkage Dendogram")
dendrogram(heirarchical_average)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Ward Linkage Dendogram")
dendrogram(heirarchical_ward)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Ward Linkage Dendogram with column 3 and 4")
dendrogram(heirarchical_ward_c3c4)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Ward Linkage Dendogram with column 1 and 3")
dendrogram(heirarchical_ward_c1c3)
plt.show()
plt.figure(figsize=(15, 9))
plt.title("Ward Linkage Dendogram with column 1 and 2")
dendrogram(heirarchical_ward_c1c2)
plt.show()

```

### Observation

#### Changing the type of linkages
We observe that on changing the type of linkages, the distance of separation between two clusters varies greatly. 
Single Linkage creates long but good separated clusters, while complete linkage creates compact but not so well separated clusters.
Ward has the highest intra cluster distance value.

#### Changing the number of features
Again, by changing the number of clusters, we observe that the clustering changes. This is because different features are not correlated with each other.
Using columns 1 and 3 creates well balanced clusters.

## Gene Dataset

```python
f2 = "../assignment-2-ksr-0011/Data/SMAI-Dataset-gene-expression/gene.csv"
gene_data = pd.read_csv(f2)
gene_data.head()
```

```python
# Dropping ID column
gene_data = gene_data.drop(['ID_REF'], axis=1)
```

#### Varying the type of linkage

```python

gene_data_ward = linkage(gene_data, method='ward')
gene_data_single = linkage(gene_data, method='single')
gene_data_complete = linkage(gene_data, method='complete')
gene_data_average = linkage(gene_data, method='average')

plt.figure(figsize=(14, 7))
plt.title("Ward Linkage Dendogram")
dendrogram(gene_data_ward)
plt.show()
plt.figure(figsize=(14, 7))
plt.title("Single Linkage Dendogram")
dendrogram(gene_data_single)
plt.show()
plt.figure(figsize=(14, 7))
plt.title("Complete Linkage Dendogram")
dendrogram(gene_data_complete)
plt.show()
plt.figure(figsize=(14, 7))
plt.title("Average Linkage Dendogram")
dendrogram(gene_data_average)
plt.show()
```

### Varying the number of features


```python
gene_data_ward_10 = linkage(gene_data.iloc[:, :10], method='ward')
gene_data_ward_9 = linkage(gene_data.iloc[:, :9], method='ward')
gene_data_ward_8 = linkage(gene_data.iloc[:, :8], method='ward')
gene_data_ward_7 = linkage(gene_data.iloc[:, :7], method='ward')
gene_data_ward_6 = linkage(gene_data.iloc[:, :6], method='ward')
gene_data_ward_5 = linkage(gene_data.iloc[:, :5], method='ward')
gene_data_ward_4 = linkage(gene_data.iloc[:, :4], method='ward')
gene_data_ward_3 = linkage(gene_data.iloc[:, :3], method='ward')
gene_data_ward_2 = linkage(gene_data.iloc[:, :2], method='ward')

plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 10 features")
dendrogram(gene_data_ward_10)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 9 features")
dendrogram(gene_data_ward_9)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 8 features")
dendrogram(gene_data_ward_8)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 7 features")
dendrogram(gene_data_ward_7)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 6 features")
dendrogram(gene_data_ward_6)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 5 features")
dendrogram(gene_data_ward_5)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 4 features")
dendrogram(gene_data_ward_4)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 3 features")
dendrogram(gene_data_ward_3)
plt.show()
plt.figure(figsize=(15, 10))
plt.title("Ward Linkage Dendogram with 2 features")
dendrogram(gene_data_ward_2)
plt.show()

```

### Changing Linkages
There was not much significance difference in the clustering when the linkage was changed. This might be because the data is well spread out and the clusters are well separated.

### Changing the number of features
The clustering changes when the number of features are changed. Increasing the number of feature used gives better and more balanced clusters.
